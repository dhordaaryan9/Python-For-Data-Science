# 1. Write a python program to sum of the first n positive integers.


```python
user = int(input("Enter any number: "))
total = 0

if user < 0:
    print("Please enter valid number.")
else:
    for i in range(1,user+1):
        total += i
    print(f"Sum of the first {user} integer is: {total}")
```

# 3. Write a Python program to count the occurrences of each word in a given sentence.

```python
text = "The cat sat on the mat with another cat"
words = text.lower().split()

word_count = {}

for i in words:
    if i in word_count:
        word_count[i] += 1
    else:
        word_count[i] = 1

print(word_count)
```

# 6. Write a Python program to find the first appearance of the substring 'not' and 'poor' from a given string, if 'not' follows the 'poor', replace the whole 'not'...'poor' substring with 'good'. Return the resulting string

```python
text = 'The lyrics are not that poor!'
# text = 'The lyrics are poor!'

not_pos = text.find('not')
poor_pos = text.find('poor')

if not_pos != -1 and poor_pos != -1 and not_pos < poor_pos:

    text = text[:not_pos] + 'good' + text[poor_pos + 4:]
    print(text)

else:
    print(text)
```

# 8. Write a Python program to check whether a list contains a sublist.

```python
main_list = [1, 2, 3, 4, 5, 6]
sub_list = [3, 4, 5]

n = len(main_list)
m = len(sub_list)
found = False

for i in range(n - m + 1):
    if main_list[i : i + m] == sub_list:
        found = True
        break 

if found:
    print("Yes, the sublist is present.")
else:
    print("No, the sublist is not there.")
```

# 13.Write a Python program to sort a dictionary (ascending /descending) by value.


```python
data = {'apple': 50, 'orange': 10, 'banana': 30}

sorted_asc = dict(sorted(data.items(), key=lambda item: item[1]))

sorted_desc = dict(sorted(data.items(), key=lambda item: item[1], reverse=True))

print("Ascending:", sorted_asc)
print("Descending:", sorted_desc)
```

# 15.Given a number n, write a python program to make and print the list of Fibonacci series up to n.
Input : n=7 Hint : first 7 numbers in the series

Expected output : First few Fibonacci numbers are 0, 1, 1, 2, 3, 5, 8, 13

```python
n = 7

fib_list = [0, 1]

for i in range(2, n + 1):

    next_value = fib_list[-1] + fib_list[-2]
    fib_list.append(next_value)

print(f"First {n} Fibonacci numbers are: {fib_list}")
```

# 16.Counting the frequencies in a list using a dictionary in Python.
Input : [1, 1, 1, 5, 5, 3, 1, 3, 3, 1,4, 4, 4, 2, 2, 2, 2]

Expected output : 1 : 5 , 2 : 4 , 3 : 3 , 4 : 3 , 5 : 2

```python
data = [1, 1, 1, 5, 5, 3, 1, 3, 3, 1, 4, 4, 4, 2, 2, 2, 2]
freq = {}

for num in data:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

# 17.Write a python program using function to find the sum of odd series and even series
Odd series: 12/ 1! +32/ 3! + 52/ 5!+……n

Even series: 22/ 2! + 42/ 4! + 62/ 6!+……n

```python
import math

def calculate_series_sums(n):
    odd_sum = 0
    even_sum = 0

    for i in range(1, n + 1, 2):
        term = (i**2) / math.factorial(i)
        odd_sum += term

    for i in range(2, n + 1, 2):
        term = (i**2) / math.factorial(i)
        even_sum += term

    return odd_sum, even_sum

limit = int(input("Enter the value of n: "))
odds, evens = calculate_series_sums(limit)

print(f"Sum of Odd series: {odds:.4f}")
print(f"Sum of Even series: {evens:.4f}")
```

# 18.Python Program to Find Factorial of Number Using Recursion


```python
num = int(input("Enter a number: "))
factorial = 1

if num < 0:
    print("Factorial does not exist for negative numbers.")

elif num == 0:
    print("The factorial of 0: 1")

else:
    for i in range(1, num + 1):
        factorial = factorial * i
        
    print(f"The factorial of {num}: {factorial}")
```

# 20.Mini project : Problem Statement : Password Generator

Make a program to generate a strong password using the input given by the user.
To generate a password, randomly take some words from the user input and then
include numbers, special characters and capital letters to generate the password.
Also, keep a check that password length is more than 8 characters.


Note: Include Exception handling wherever required. Also, make a ‘User’ class
and store the details like user id, name and password of each user as a tuple.

```python
import random
import string

class User:
    def __init__(self, uid, name, pwd):
        self.data = (uid, name, pwd)

def generate_pass(words):
    try:
        if len(words) < 2:
            raise ValueError("Need at least 2 words")
        
        base = random.choice(words) + random.choice(words)
        extra = random.choice(string.digits) + random.choice("!@#$")
        password = base.capitalize() + extra
        
        if len(password) < 8:
            password += "12345"
        return password
    except Exception as e:
        return f"Error: {e}"

u_id = input("ID: ")
name = input("Name: ")
words = input("Enter words: ").split()

pwd = generate_pass(words)

if "Error" not in pwd:
    userobj = User(u_id, name, pwd)
    print(f"Stored Tuple: {userobj.data}")
else:
    print(pwd)
```
